```python
prod.head(2)
```

```python
prod.info()
```

```python
prod['Attempted time']=pd.to_datetime(prod['Attempted time'])
```

```python
prod.info()
```

```python
prod.head(10)
```

```python
list_days=[]
for i in range(prod.shape[0]):
    list_days.append(prod['Attempted time'][i].month)
```

```python
list_days
```

```python
prod['Day']=list_days
```

```python
prod.head(4)
```

```python
prod2 = prod.groupby(['driver','Day']).count()
```
